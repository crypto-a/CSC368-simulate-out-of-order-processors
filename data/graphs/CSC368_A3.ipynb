# GRAPH 1: IPC Comparison Across Designs and Workloads

# Task
Generate a grouped bar chart using `plotly.express` to compare IPC across different designs and workloads, coloring the bars by `design` and setting `workload` on the x-axis and `ipc` on the y-axis, with `barmode='group'`. Define a custom color map, prepare legend labels to include credit costs for each design, and apply chart customizations including the title, x-axis and y-axis labels, figure width and height, horizontal gridlines, and x-axis label rotation. ## import_libraries

### Subtask:
Import `pandas` for data manipulation and `plotly.express` for plotting.


**Reasoning**:
Import the `plotly.express` library as `px` for plotting, as specified in the instructions.



```python
import plotly.express as px
print("plotly.express imported successfully as px.")
```

## define_chart_parameters

### Subtask:
Define the color map and prepare the legend labels including credit costs for each design based on the specifications.


**Reasoning**:
First, I will extract unique designs and their credit costs from the dataframe. Then, I will create a dictionary to map each design to its credit cost, formatted for legend labels, and define a custom color map for each design. These will be used for the upcoming visualization.



```python
unique_designs_credits = df[['design', 'credits']].drop_duplicates().set_index('design')['credits'].to_dict()

design_credit_map = {
    design: f"{design} (credits: {credits})"
    for design, credits in unique_designs_credits.items()
}

color_map = {
    'design_a': '#1f77b4',
    'design_b': '#ff7f0e',
    'design_c': '#2ca02c'
}

print("Unique designs and their credit costs:", unique_designs_credits)
print("Design credit map for legend labels:", design_credit_map)
print("Custom color map:", color_map)
```

**Reasoning**:
The previous `color_map` is incomplete as it does not include a color for 'design_d', which was identified in the `unique_designs_credits`. I need to update the `color_map` to ensure all unique designs have a corresponding color, as specified in the instructions.



```python
color_map = {
    'design_a': '#1f77b4',
    'design_b': '#ff7f0e',
    'design_c': '#2ca02c',
    'design_d': '#d62728' # Adding a color for design_d
}

print("Updated custom color map:", color_map)
```

**Reasoning**:
Now that the color map and legend labels are defined, I will generate the grouped bar chart using `plotly.express` as specified in the main task. I will use the `df` DataFrame, `workload` on the x-axis, `ipc` on the y-axis, color by `design`, and set `barmode='group'`. I will also apply the custom color map and legend labels, and then add chart customizations like title, axis labels, figure size, gridlines, and x-axis label rotation.



```python
fig = px.bar(df,
             x='workload',
             y='ipc',
             color='design',
             barmode='group',
             color_discrete_map=color_map,
             labels={
                 'workload': 'Workload',
                 'ipc': 'IPC (Instructions Per Cycle)',
                 'design': 'Design'
             },
             title='IPC Comparison Across Designs and Workloads',
             category_orders={'design': list(design_credit_map.keys())} # Ensure consistent order
            )

# Update legend labels to include credit costs
for i, design in enumerate(design_credit_map.keys()):
    fig.data[i].name = design_credit_map[design]

# Customize layout
fig.update_layout(
    xaxis_title='Workload',
    yaxis_title='IPC',
    width=1000, # Set figure width
    height=600, # Set figure height
    yaxis=dict(
        gridcolor='lightgray', # Horizontal gridlines
        gridwidth=1
    ),
    xaxis=dict(
        tickangle=-45 # Rotate x-axis labels
    )
)

fig.show()
```

## Summary:

### Data Analysis Key Findings

*   The analysis successfully compared Instruction Per Cycle (IPC) values across four distinct designs (`design_a`, `design_b`, `design_c`, `design_d`) against various workloads.
*   A custom color map was applied to visually differentiate the designs in the chart.
*   Legend labels were enhanced to include the credit cost for each design, providing additional context (e.g., "design\_a (credits: 820)").
*   A grouped bar chart was generated, effectively visualizing IPC performance for each design under different workloads, allowing for direct comparison.

### Insights or Next Steps
